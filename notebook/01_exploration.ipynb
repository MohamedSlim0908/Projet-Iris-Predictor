# Iris Predictor Exploration
Rapide exploration du dataset Iris pour comprendre sa structure et calibrer le pipeline.

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")

def prepare_iris_frame():
    iris = load_iris(as_frame=True)
    frame = iris.frame.copy()
    frame.columns = [col.replace(" (cm)", "").replace(" ", "_") for col in frame.columns]
    frame["target_name"] = frame["target"].map(dict(enumerate(iris.target_names)))
    return frame, iris.target_names

df, target_names = prepare_iris_frame()
feature_cols = [col for col in df.columns if col not in {"target", "target_name"}]
feature_cols
```

```python
df.head()
```

```python
df[feature_cols].describe().T
```

```python
ax = df[feature_cols].plot(kind="hist", alpha=0.6, bins=20, figsize=(8, 5))
ax.set_xlabel("cm")
ax.set_title("Distribution des caractéristiques")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
for name in target_names:
    subset = df[df["target_name"] == name]
    ax.scatter(subset["petal_length"], subset["petal_width"], label=name, alpha=0.7)
ax.set_xlabel("petal_length (cm)")
ax.set_ylabel("petal_width (cm)")
ax.set_title("Séparation par mesures de pétales")
ax.legend()
plt.show()
```

Les mesures de pétales offrent la meilleure séparation entre espèces, ce qui valide l'usage d'un modèle linéaire régularisé combiné à une normalisation.
